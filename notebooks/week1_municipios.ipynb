```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
df_deathsm = pd.read_csv('Casos_Diarios_Municipio_Defunciones_20210621.csv')
a, b = df_deathsm.shape
df_deathsm.head(10)
```

```python
df_deathsm['total']= df_deathsm.loc[:,'20-01-2020':'20-06-2021'].sum(axis=1)
df_deathsm.head(10)
```

```python
df_deathsm.sort_values('total', inplace=True, ascending=False)
df_deathsm.head()
```

```python
df_deathsm['rates'] = (df_deathsm['total']/df_deathsm['poblacion'])*100000
df_deathsm.head()
```

```python
df_deathsm['rates'] = (df_deathsm['total']/df_deathsm['poblacion'])*100000
df_deathsm.head()
```

```python
plt.figure(figsize=(14,10))
sns.barplot(x="total", y="nombre", data=df_deathsm.nlargest(10, 'poblacion'))
sns.set_context('paper', font_scale=2)
```

```python
df_deathsm['total60']= df_deathsm.iloc[:, b -59 : b ].sum(axis=1)
df_deathsm['rates60'] = (df_deathsm['total60']/df_deathsm['poblacion'])*100000
df_deathsm
```

```python
plt.figure(figsize=(14,10))
sns.barplot(x="rates60", y="nombre", data=df_deathsm.nlargest(10, 'rates60'))
sns.set_context('paper', font_scale=2)
```
