```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
df_deathse = pd.read_csv('Casos_Diarios_Estado_Nacional_Defunciones_20210620.csv')
```

```python
df_deathse = df_deathse.query("nombre != 'Nacional'").copy()
df_deathse = df_deathse.query("nombre != 'Nacional'").copy()
a, b = df_deathse.shape
df_deathse.head()
```

```python
df_deathse['total']= df_deathse.loc[:,'20-01-2020':'19-06-2021'].sum(axis=1)
```

```python
df_deathse.head(10)
```

```python
df_deathse.sort_values('total', inplace=True, ascending=False)
df_deathse.head()
```

```python
df_deathse.shape
```

```python
pd.set_option('display.max_rows', 33)
pd.set_option('display.max_columns', 541)
```

```python
df_deathse.nlargest(10, 'poblacion')
```

```python
df_deathse['rates'] = (df_deathse['total']/df_deathse['poblacion'])*100000
df_deathse.head()
```

```python
plt.figure(figsize=(14,10))
sns.barplot(x="total", y="nombre", data=df_deathse.nlargest(10, 'poblacion'))
sns.set_context('paper', font_scale=2)
```

```python
df_deathse['total60']= df_deathse.iloc[:, b -59 : b ].sum(axis=1)
df_deathse['rates60'] = (df_deathse['total60']/df_deathse['poblacion'])*100000
df_deathse
```

```python
plt.figure(figsize=(14,10))
sns.barplot(x="rates60", y="nombre", data=df_deathse.nlargest(10, 'rates60'))
sns.set_context('paper', font_scale=2)
```
